```python
#import dependencies
import pandas as pd
from citipy import citipy
import random
import config as cfg
import gmap as gmp
import requests
import matplotlib.pyplot as plt
from scipy.stats import linregress
import gmaps
import json
```

```python
weather_data = pd.read_csv("Analysis\weather_df.csv")
# gmaps_api = "AIzaSyC8jKYVQIsB3o3jJhc-eIFV4Cqh6VRSAXc"
# Access maps with unique API key
gmaps.configure(api_key={gmp.gkey})

# Store 'Lat' and 'Lng' into  locations 
locations = weather_data[["Latitude", "Longitude"]].astype(float)

# Convert Poverty Rate to float and store
# HINT: be sure to handle NaN values

humidity_rate = weather_data["Humidity"].astype(float)
# Customize the size of the figure
figure_layout = {
    'width': '500px',
    'height': '300px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)

max_intense = int(weather_data["Humidity"].max())
heat_layer = gmaps.heatmap_layer(locations, weights=humidity_rate, 
                                 dissipating=True, max_intensity=max_intense,
                                 point_radius = 10)

fig.add_layer(heat_layer)

display(fig)
```

```python
# Narrow down the DataFrame to find your ideal weather condition.:
ideal_weather_data = weather_data[(weather_data["Temperature"] > int(60)) & 
                                  (weather_data["Temperature"] < int(70)) &
                                  (weather_data["Wind"] < int(10)) &
                                  (weather_data["Clouds"] == int(0)) 
                                 ] 
display(ideal_weather_data)
```

```python
gmaps_api = "AIzaSyC8jKYVQIsB3o3jJhc-eIFV4Cqh6VRSAXc"

hotel_list = []

target_search = "Hotel"
target_radius = 5000
target_type = "lodging"

for x in range(len(ideal_weather_data)):
    lat = ideal_weather_data['Latitude'].iat[x]
    lng = ideal_weather_data['Longitude'].iat[x]
#     lat = "43.6187102"
#     lng = "-116.2146068"
#     lat = "41.113941"
#     lng = "37.262046"
    
    params = {
        "location": f"{lat},{lng}",
        "keyword": target_search,
        "radius": target_radius,
        "type": target_type,
        "key": gmaps_api
    }
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    response = requests.get(base_url, params=params)
    jsn = response.json()
    print(jsn)

    
    
    try:
        hotel_list.append(jsn['results'][0]['name'])
    except:
        hotel_list.append("")
ideal_weather_data["Hotel Name"] = hotel_list
ideal_weather_data = ideal_weather_data.dropna(how='any')
ideal_weather_data.head()


```

```python

info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>

</dl>
"""

# <dt>City</dt><dd>{City}</dd>
# <dt>Country</dt><dd>{Country}</dd>

locations = ideal_weather_data[["Latitude", "Longitude"]]

markers = gmaps.marker_layer(locations)
# Create hotel symbol layer
hotel_layer = gmaps.marker_layer(
    locations,info_box_content=[info_box_template.format(**row) for index, row in ideal_weather_data.iterrows()]
)

# Add layer
fig.add_layer(markers)
fig.add_layer(hotel_layer)

fig
```
